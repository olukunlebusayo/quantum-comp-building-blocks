The goal of these codes is to show four different ways by which quantum gates are represented:
    #1. by writing out an expression e.g. for X gate we have:
    X|0> = |1>
    X|1> = |0>
    
    #2. by writing in matrix form (linear algebra). X gate in this case
    (2by2)[[0, 1], (2by1)[1,0] = (2by1)[0, 1]
          [1, 0]]
    
    #3. visual representation by plotting the result of the gate (X gate in this case) on bloch vector
    plot_bloch_multivector()
    
    #. running the cct thru a measurement and seeing what the measurement outcomes are.
    plot_histogram()

```python
from qiskit import *
```

```python
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
#statevector is the vector that describes the quantum state of our qubit
circuit = QuantumCircuit(1, 1) #quantum cct with 1 qbit and 1 classical bit
circuit.x(0) #one operation on the quantume cct which is apply x to it
```

```python
#take the quantum cct and simulate it

simulator = Aer.get_backend('statevector_simulator')
```

```python
#execute the simulator using the simulator backend and store the result in a variable result
result = execute(circuit, backend=simulator).result()
```

```python
statevector = result.get_statevector()
```

```python
print(statevector)
```

```python
#the quantum state got above shows that X applied on state 0 is 1 plus zero imaginary component

#let's draw to see this in schematic form

%matplotlib inline

circuit.draw(output = 'mpl')
```

```python
#additional ways to discuss quantum state
#operations on quantum state is alwasys referred to by many as rotations on the sphere

#1. this is going to be more visual and it's favorite of many

#in the plot below, what X gate has done is to take the quantum state from state 0 to 1

plot_bloch_multivector(statevector)
```

```python
#we can measure the output of the circuit

circuit.measure([0], [0]) #this is taking the measurement of the qr and storing it in cr
backend = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = backend, shots=1024).result()
counts = result.get_counts()

from qiskit.tools.visualization import plot_histogram

plot_histogram(counts)
```

```python
#let's get the matrix representation of the cct
#X gate is represented by matrix [[0,1], [1,0]]

circuit = QuantumCircuit(1, 1)
circuit.x(0)
simulator = Aer.get_backend('unitary_simulator')
result = execute(circuit, backend=simulator).result()
unitary = result.get_unitary()
print(unitary)
```
